### A1 — Max, Min, Frequency using only loops (no built-in functions)

No `max()`, `min()`, `len()` inside these functions — only `for` loops and basic comparisons.

```
def find_max(lst):
    current_max = lst[0]
    for item in lst:
        if item > current_max:
            current_max = item
    return current_max


def find_min(lst):
    current_min = lst[0]
    for item in lst:
        if item < current_min:
            current_min = item
    return current_min


def count_frequency(lst):
    freq = {}
    for item in lst:
        if item in freq:
            freq[item] = freq[item] + 1
        else:
            freq[item] = 1
    return freq


# ── Test ──────────────────────────────────────────────────────────────────
numbers = [4, 7, 2, 9, 1, 5, 9, 3, 7, 7, 2, 4, 4, 4]

print(f"List:      {numbers}")
print(f"Maximum:   {find_max(numbers)}")
print(f"Minimum:   {find_min(numbers)}")
print()
print("Frequency count:")
freq_result = count_frequency(numbers)
for val in sorted(freq_result):
    print(f"  {val} appears {freq_result[val]} time(s)")
```

### A2 — While loop: reverse a number and check palindrome

```
def reverse_number(n):
    """
    Reverses digits of n using a while loop.
    Example: 123 -> 321
    Works by repeatedly taking the last digit (n % 10)
    and building the reversed number digit by digit.
    """
    original  = n
    reversed_n = 0
    while n > 0:
        last_digit  = n % 10          # pluck the last digit
        reversed_n  = reversed_n * 10 + last_digit   # append to result
        n           = n // 10         # chop off the last digit
    return reversed_n


def is_palindrome(n):
    """
    A number is a palindrome if it equals its reverse.
    Example: 121 -> reversed is 121 -> palindrome
             123 -> reversed is 321 -> not a palindrome
    """
    return n == reverse_number(n)


# ── Test ──────────────────────────────────────────────────────────────────
test_nums = [123, 121, 1001, 12321, 456, 9009]

for num in test_nums:
    rev = reverse_number(num)
    pal = is_palindrome(num)
    print(f"  {num}  ->  reversed: {rev:6d}  |  palindrome: {pal}")
```

### A3 — Tuples & Dictionaries: convert list of tuples + find key with highest value

```
def tuples_to_dict(tuple_list):
    """
    Converts [(key, value), ...] into a dict without using dict() or {}.
    Builds an empty dict then inserts one entry at a time.
    """
    result = {}
    for pair in tuple_list:
        key   = pair[0]
        value = pair[1]
        result[key] = value
    return result


def key_with_highest_value(d):
    """
    Finds the key whose value is largest — using only a loop.
    No max(), no sorted().
    """
    best_key   = None
    best_value = None
    for key in d:
        if best_value is None or d[key] > best_value:
            best_key   = key
            best_value = d[key]
    return best_key, best_value


# ── Test ──────────────────────────────────────────────────────────────────
pairs = [("apple", 50), ("banana", 30), ("mango", 90), ("grape", 70), ("orange", 45)]

converted = tuples_to_dict(pairs)
print(f"Input tuples:  {pairs}")
print(f"Converted dict: {converted}")

best_k, best_v = key_with_highest_value(converted)
print(f"Key with highest value: '{best_k}' = {best_v}")
```

### A4 — `*args`: sum and average of multiple numbers (no `sum()` inside)

```
def sum_and_average(*args):
    """
    Accepts any number of arguments.
    Computes total with a for loop (no sum()), then divides for average.
    """
    total = 0
    count = 0
    for num in args:
        total = total + num
        count = count + 1

    if count == 0:
        return 0, 0

    average = total / count
    return total, average


# ── Test ──────────────────────────────────────────────────────────────────
t1, a1 = sum_and_average(10, 20, 30)
print(f"sum_and_average(10, 20, 30)          -> total={t1},  avg={a1:.2f}")

t2, a2 = sum_and_average(5, 15, 25, 35, 45, 55)
print(f"sum_and_average(5,15,25,35,45,55)    -> total={t2},  avg={a2:.2f}")

t3, a3 = sum_and_average(100)
print(f"sum_and_average(100)                 -> total={t3},  avg={a3:.2f}")
```

### A5 — `**kwargs`: accept student names and marks, return the top scorer

```
def top_student(**kwargs):
    """
    Accepts student_name=marks as keyword arguments.
    Returns the name and mark of the student with the highest score.
    Uses only a for loop — no max() or sorted().
    """
    best_name  = None
    best_mark  = None

    for name in kwargs:
        mark = kwargs[name]
        if best_mark is None or mark > best_mark:
            best_name = name
            best_mark = mark

    return best_name, best_mark


# ── Test ──────────────────────────────────────────────────────────────────
result_name, result_mark = top_student(
    Ravi=78, Priya=91, Arjun=55, Sunita=63,
    Deepak=82, Kavya=95, Nikhil=70
)
print("Student marks:")
all_marks = dict(Ravi=78, Priya=91, Arjun=55, Sunita=63,
                 Deepak=82, Kavya=95, Nikhil=70)
for name, mark in all_marks.items():
    star = " ← TOP" if name == result_name else ""
    print(f"  {name:10s}: {mark}{star}")
print(f"Top student: {result_name} with {result_mark} marks")
```

## 🚀 Part B — Stretch: Custom `Vector` class with Dunder Methods

```
class Vector:
    """
    Represents a mathematical vector of any dimension.
    Stored as a tuple so it is immutable by default.

    Dunder methods implemented:
      __init__   : store the components as a tuple
      __repr__   : readable output in Vector(x, y, z) style
      __add__    : element-wise addition
      __sub__    : element-wise subtraction
      __mul__    : scalar multiplication (Vector * scalar)
      __rmul__   : reverse scalar multiplication (scalar * Vector)
      __eq__     : equality check
      __len__    : number of dimensions
    """

    def __init__(self, *components):
        self.components = tuple(components)

    def __repr__(self):
        parts = ", ".join(str(c) for c in self.components)
        return f"Vector({parts})"

    def __add__(self, other):
        if len(self.components) != len(other.components):
            raise ValueError("Vectors must have the same number of dimensions to add.")
        new_parts = []
        for i in range(len(self.components)):
            new_parts.append(self.components[i] + other.components[i])
        return Vector(*new_parts)

    def __sub__(self, other):
        if len(self.components) != len(other.components):
            raise ValueError("Vectors must have the same number of dimensions to subtract.")
        new_parts = []
        for i in range(len(self.components)):
            new_parts.append(self.components[i] - other.components[i])
        return Vector(*new_parts)

    def __mul__(self, scalar):
        new_parts = []
        for c in self.components:
            new_parts.append(c * scalar)
        return Vector(*new_parts)

    def __rmul__(self, scalar):
        # allows scalar * Vector in addition to Vector * scalar
        return self.__mul__(scalar)

    def __eq__(self, other):
        return self.components == other.components

    def __len__(self):
        count = 0
        for _ in self.components:
            count += 1
        return count
```

```
# ── Test the Vector class ─────────────────────────────────────────────────
v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)
v3 = Vector(0, -1, 2)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v3 = {v3}")
print()
print(f"v1 + v2        = {v1 + v2}")
print(f"v2 - v1        = {v2 - v1}")
print(f"v1 * 3         = {v1 * 3}")
print(f"2 * v2         = {2 * v2}")
print(f"v1 + v3        = {v1 + v3}")
print(f"v1 == v1       = {v1 == v1}")
print(f"v1 == v2       = {v1 == v2}")
print(f"len(v1)        = {len(v1)}")
print()

# chain of operations
result = (v1 + v2) * 2 - v3
print(f"(v1 + v2) * 2 - v3 = {result}")

# 2D vectors
a = Vector(3, 4)
b = Vector(1, -2)
print(f"2D: {a} + {b} = {a + b}")
print(f"2D: {a} * 5     = {a * 5}")

# dimension mismatch
try:
    bad = Vector(1, 2) + Vector(1, 2, 3)
except ValueError as e:
    print(f"Dimension mismatch caught: {e}")
```

### Q1 — `*args` vs `**kwargs`

```
print("""
*args   (positional arguments):
  Collects any number of POSITIONAL arguments into a TUPLE.
  Use when you don't know how many values the caller will pass,
  but those values have no names -- they are just a sequence.

  Example:
    def add(*args):
        total = 0
        for n in args: total += n
        return total
    add(1, 2, 3)       # args = (1, 2, 3)
    add(10, 20)        # args = (10, 20)

  Real use cases:
    - print() itself uses *args to accept any number of values
    - A logger function that accepts any number of messages
    - Math functions where input count varies (sum, product)

**kwargs  (keyword arguments):
  Collects any number of NAMED arguments into a DICT.
  Use when arguments have meaningful names and you want to
  access them by name inside the function.

  Example:
    def register(**kwargs):
        for key, val in kwargs.items():
            print(f'{key}: {val}')
    register(name='Ravi', age=25, city='Mumbai')

  Real use cases:
    - Configuration functions (pass any settings by name)
    - Decorator factories that forward options
    - Database record builders: insert_row(**{'col1': v1, 'col2': v2})

Using both together:
  def func(*args, **kwargs):
      ...
  args are always positional, kwargs are always named.
  Order must be: regular params -> *args -> **kwargs
""")
```

### Q2 — `Student` class with grade and `__str__`

```
class Student:
    """
    Represents a student with name and marks.

    Methods:
      calculate_grade() -- returns A, B, or C based on marks
      __str__           -- readable one-line summary
    """

    def __init__(self, name, marks):
        self.name  = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 60:
            return "B"
        else:
            return "C"

    def __str__(self):
        grade = self.calculate_grade()
        return f"Student: {self.name:<12s} | Marks: {self.marks:3d} | Grade: {grade}"


# ── Test ──────────────────────────────────────────────────────────────────
students = [
    Student("Ravi",   78),
    Student("Priya",  91),
    Student("Arjun",  55),
    Student("Sunita", 63),
    Student("Deepak", 82),
    Student("Kavya",  47),
]

print("Student Report:")
print("-" * 45)
for s in students:
    print(s)

print()
# demonstrate that grade method works standalone
top = students[1]
print(f"Top scorer: {top.name}, grade = {top.calculate_grade()}")
```

### Q3 — Dunder methods: what they are and why they matter

```
print("""
DUNDER METHODS (also called magic methods or special methods):

  Dunder = double underscore on both sides: __name__
  They are Python's hook system -- you define them in your class
  and Python calls them automatically in response to built-in
  operations or syntax.

  Why they matter:
    1. They let your custom objects BEHAVE LIKE built-in types.
       Your class can support +, -, *, len(), print(), ==, [] etc.
    2. They make your objects work seamlessly with Python idioms
       (for loops, f-strings, context managers, comparison operators).
    3. They follow a consistent protocol -- any class with __iter__
       works in a for loop; any class with __len__ works with len().

Three examples:

  __str__(self)
    Called by str(obj) and by print(obj).
    Should return a human-readable string.
    Example: print(student) -> 'Student: Ravi | Marks: 78 | Grade: B'

  __add__(self, other)
    Called when you write obj1 + obj2.
    Example: Vector(1,2) + Vector(3,4) -> Vector(4,6)
    Without this, Python raises TypeError: unsupported operand type(s) for +

  __len__(self)
    Called by len(obj).
    Example: len(Vector(1,2,3)) -> 3
    Also used implicitly in truth testing: if obj: -> calls __bool__
    or falls back to __len__ (empty = False if __len__ returns 0).

Other common ones:
  __repr__  : unambiguous representation, used in REPL and debugging
  __eq__    : obj1 == obj2
  __lt__    : obj1 < obj2  (enables sorting)
  __getitem__: obj[key]    (enables indexing)
  __iter__  : for item in obj  (enables iteration)
  __enter__/__exit__: with statement (context managers)
""")
```
